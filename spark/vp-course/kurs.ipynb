## when

```python
from pyspark.sql.functions import when
```

```python
salary = spark.createDataFrame([
    (1,"Anna",1000),
    (3,"John",2000),
    (4, "Joe",3000)
], ["user_id", "name","salary"])
```

```python
salary.withColumn("newCol", 
                  when(col("salary") > 2500, "1")
                  .when(col("salary")>= 1000, "2")
                  .otherwise("NN")).show()
```

## map

```python
salary
```

```python
new_schema = StructType(
    [
    StructField("body",
                MapType(StringType(), StringType()),
                False)
    ]
)
```

```python
from pyspark.sql import Row

def map_to_json(r):
    return Row(
        body={
            "id": r.user_id,
            "name": r.name,
            "salary": str(r.salary * 12)
        }
    )
```

```python
emps = salary.rdd.map(lambda r: map_to_json(r)).toDF(new_schema)
emps.show(truncate=False)
```

## flatmap

```python
rdd = spark.sparkContext.parallelize(["apple,banana", "cherry,grape"])

print("map:")
rdd.map(lambda x: x.split(",")).collect()

print("flatMap:")
rdd.flatMap(lambda x: x.split(",")).collect()
```
