```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../sales.csv", parse_dates=["date"])
df["total"] = df["quantity"] * df["price"]
df
```

```python
product_summary = df.groupby("product", as_index=False)["total"].sum()
```

```python
sns.barplot(data=product_summary, x="product", y="total")
plt.title("Total sales per product")
plt.show()
```

```python
import matplotlib.dates as mdates

product = "apple"
filtered = df[df["product"] == product]

daily = filtered.groupby("date", as_index=False)["total"].sum()

sns.lineplot(data=daily, x="date", y="total", marker="o")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.title(f"Daily sales for {product}")
plt.xticks(rotation=45)
plt.show()
```
