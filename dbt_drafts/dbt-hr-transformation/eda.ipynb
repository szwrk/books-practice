## Departments CSV

**Setup Pandas + DuckDb**

```python
import pandas as pd
import duckdb as db

df = pd.read_csv("hr_dwh/seeds/departments_dbt.csv")
```

```python
db.register('dep',df)
```

```python
db.sql("DESCRIBE SELECT * FROM read_csv_auto('hr_dwh/seeds/departments_dbt.csv')")
```

**Table preview**

```python
db.sql("SELECT * FROM dep LIMIT 100").df()
```

**test**

```python
db.sql('''
       SELECT 
         COALESCE(DEPARTMENT_ID, -1) as DEPARTMENT_ID
        ,COALESCE(DEPARTMENT_NAME, 'Unknown') AS DEPARTMENT_NAME
        ,COALESCE(MANAGER_ID, -1) AS MANAGER_ID
        ,COALESCE(LOCATION_ID, -1) AS LOCATION_ID
       FROM dep 
       LIMIT 100
       ''').df()
```

**Count rows**

```python
db.sql("SELECT COUNT(*) FROM dep").df()
```

```python
db.sql("SELECT count(*), count(distinct DEPARTMENT_ID) FROM dep").df()
```

```python
db.sql(
'''
SELECT 
    department_id AS department_id
  , NULLIF(department_name, '') AS department_name
  , manager_id AS manager_id
  , location_id AS location_id
  , COUNT(*) AS ile
FROM dep
GROUP BY 1, 2, 3, 4
HAVING COUNT(*) > 1
''').df()
```

## Stage DB

```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
load_dotenv("prod.env", override=True)


user = os.getenv("ORACLE_USER")
password = os.getenv("ORACLE_PASSWORD")
host = os.getenv("ORACLE_HOST")
service = os.getenv("ORACLE_SERVICE")

conn_str = f"oracle+oracledb://{user}:{password}@{host}:1521/{service}"
print(conn_str)
print("user:", os.getenv("ORACLE_USER"))

engine = create_engine(conn_str)

query = """
SELECT * FROM dual
"""

df = pd.read_sql(query, engine)
df.head()
```

```python
query = """
SELECT COUNT(*), COUNT(DISTINCT EMPLOYEE_ID) FROM EMPLOYEES_RAW
"""
df = pd.read_sql(query, engine)
df.head()
```

```python
query = """
SELECT EMPLOYEE_ID, COUNT(*)
FROM EMPLOYEES_RAW
GROUP BY EMPLOYEE_ID
HAVING COUNT(*) > 1
"""
df = pd.read_sql(query, engine)
df.head()
```
